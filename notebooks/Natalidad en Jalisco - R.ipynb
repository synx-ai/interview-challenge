## Proyecciones de Natalidad en Jalisco (R)

El objetivo de este proyecto, es predecir la natalidad en el Estado de Jalisco para los años 2019 al 2021, utilizando los datos anuales del 2010 al 2018. Así como ofrecer una evaluación de los resultados.

```r
library("RSQLite")

con <- dbConnect(drv=RSQLite::SQLite(), dbname="data/natalidad_en_mexico.sqlite")

df <- dbGetQuery(conn=con, statement="SELECT * FROM nacimientos")
names(df) <- make.names(names(df),unique = TRUE)

head(df, 5)
```

```r
transform(df, Total = as.integer(Total), 
            Hombres = as.integer(Hombres),
            Mujeres = as.integer(Mujeres),
    No.Especificado = as.integer(No.Especificado),
                Año = as.integer(Año))
```
